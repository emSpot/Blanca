## redirecting

- (>) : we want to output, print to the screen and save the output replacing the content

- (>>) : we want to save to an existing file or append

- (<) : if we want to send data the other way

https://www.digitalocean.com/community/tutorials/an-introduction-to-linux-i-o-redirection

```python
!cat tests.txt
```

```python
!ls >tests.txt
```

```python
!cat tests.txt
```

```python
!wc -l tests.txt >tests.txt
```

```python
!wc -l tests.txt>test.txt
```
